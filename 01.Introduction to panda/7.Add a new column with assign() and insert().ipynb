```python
import pandas as pd
```

```python
# reading the CSV file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1 Adding a new column

## 1.1 assign()

When to use assign?
- Add multiple columns in a single line of code
- When you need to overwrite the values of an existing columns (best practice)

It returns a new object (a copy) with all the orginal columns in addition to the new ones

```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(0, 100, size=len(df_exams))
score2 = np.random.randint(0, 100, size=len(df_exams))
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index=np.arange(len(df_exams)))
series2 = pd.Series(score2, index=np.arange(len(df_exams)))
```

```python
# using assign() to add multiple columns
df_exams = df_exams.assign(score1=series1, score2=series2)
df_exams.head()
```

## 1.2 insert()

insert a new column at a specific position or index

```python
# using insert() to add a new column at a specific position
df_exams.insert(1, 'test', series1)
df_exams.head()
```
