```python
import pandas as pd
import numpy as np
import random
```

```python
# reading the CSV file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1 Creating Index

```python
# creating non repetitive values for the index
new_index = np.arange(0,1000)
```

```python
# shuffling the index
random.shuffle(new_index)
new_index[:5]
```

```python
# create a new column with new index
df_exams['new index'] = new_index
df_exams.head()
```

# 2 Set index

```python
# set new_index column as index
df_exams.set_index('new index', inplace=True)
df_exams.head()
```

# 3 Sort Index

```python
# sort dataframe by new index
df_exams.sort_index(inplace=True)
df_exams.head()
```

```python
# sort descending by new index
df_exams.sort_index(ascending=False, inplace=True)
df_exams.head()
```
