## 1.2 Adding a new column with an array
The dataframe has 1000 rows, so we need to create an array with 1000 elements

```python
# importing numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_scores = np.arange(1000)
```

```python
# length of the array
print(len(language_scores))
```

```python
# adding a new column to dataframe with an array
df_exams['language score'] = language_scores
df_exams.head()
```

```python
# create a random integer number between 1 and 100
int_language_score = np.random.randint(1, 100, 1000)
```

```python
# min value inclusive, max value exclusive
min(int_language_score), max(int_language_score)
```

```python
# adding a new column to dataframe with an array
df_exams['language score'] = int_language_score
df_exams.head()
```

```python
# create random float numbers between 1 and 100
float_language_score = np.random.uniform(1, 100, size=1000)
float_language_score[:10]  # display first 10 elements
```
