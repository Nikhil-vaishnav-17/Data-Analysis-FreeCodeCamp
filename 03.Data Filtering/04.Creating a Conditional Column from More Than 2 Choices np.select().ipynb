## Creating a Conditional Column from More Than 2 Choices: np.select()

```python
import numpy as np
```

```python
# create an array based on multiple place tiers (+2 choices)

# create "conditions" and "values" variables
conditions = [
    df_laptops['Price_euros'] > 3000,
    (df_laptops['Price_euros'] <= 3000) & (df_laptops['Price_euros'] > 2000),
    (df_laptops['Price_euros'] <= 2000) & (df_laptops['Price_euros'] > 800),
    df_laptops['Price_euros'] <= 800
]

values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set it to new column
df_laptops['Price_tiers'] = np.select(conditions, values, default='Other')
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "price_tiers" column
df_laptops['Price_tiers'].value_counts()
```

```python
# create an array based on multiple screen size (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small < 12

# create conditions and values variables
conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] <= 16) & (df_laptops['Inches'] > 14),
    (df_laptops['Inches'] <= 14) & (df_laptops['Inches'] > 12),
    df_laptops['Inches'] <= 12
]

values = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
# set it to a new column
df_laptops['Screen_size'] = np.select(conditions, values, default='Other')
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "screen_size" column
df_laptops['Screen_size'].value_counts()
```
