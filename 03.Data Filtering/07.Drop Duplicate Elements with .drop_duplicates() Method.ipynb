```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
df_laptops.head(3)
```

# 1 drop_duplicates()

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# sort dataframe asending by company and price
# cheapest first and most expensive last
df_laptops.sort_values(['Company', 'Price_euros'], inplace=True)
df_laptops
```

```python
# cheapest: keep='first' 
df_laptops.drop_duplicates(['Company'],keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive: keep='last' 
df_laptops.drop_duplicates(['Company'],keep='last', ignore_index=True)[['Company', 'Price_euros']]
```
