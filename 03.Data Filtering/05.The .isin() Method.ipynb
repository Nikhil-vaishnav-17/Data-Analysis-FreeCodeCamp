```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops
```

# 1 isin(): Single Filtering

```python
# select Apple or HP laptops
df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframe
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])]
```

# 2 isin(): Multiple Filtering

```python
# find notebook or ultrabooks from Apple or HP
filter1 = df_laptops['TypeName'].isin(['Notebook', 'Ultrabook'])
filter2 = df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# filter dataframe
df_laptops[filter1 & filter2].value_counts(['Company', 'TypeName'])
```
